```python
import numpy as np
from makecorner import corner
```

```python
# Generate some mock data
xs = np.random.normal(size=1000)
ys = np.random.normal(size=1000, scale = 0.54)
zs = np.random.normal(size=1000)

data = {
    'x':{'data': xs, 'plot_bounds':(-3, 3), 'label':r'$x$'},
    'y':{'data': ys, 'plot_bounds':(-3, 3), 'label':r'$y$'},
    'z':{'data': zs, 'plot_bounds':(-3, 3), 'label':r'$z$'}
}
```

## Generate a plot with default settings

```python
fig = corner(data)
```

## Add contour levels

```python
fig = corner(data, contour_levels=(0.5, 0.95))
```

## Adjust figure size, turn off credible intervals, adjust number of bins, and change colors

```python
fig = corner(data, color='#e31a1c', figsize=(4,4), show_bounds=False, bins=30)
```
